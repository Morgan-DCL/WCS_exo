### Q3 : we want to know if all the movies are with good rating, good popularity and hight vote count. to explore that calculate min, max and mean for popularity, vote_average, vote_count

solution be like

image.png

```python
min_max_votes_ =  [film["vote_average"] for film in data]
votes_count_ =  [film["vote_average"] for film in data]
popularity_ = [film["popularity"] for film in data]

pop = {
    "pop_min" : min(popularity_),
    "pop_max" : max(popularity_),
    "pop_mean" : sum(popularity_) / len(popularity_),
    "votes_min" : min(votes_count_),
    "votes_max" : max(votes_count_),
    "votes_mean" : sum(votes_count_) / len(votes_count_),
    "min_max_min" : min(min_max_votes_),
    "min_max_max" : max(min_max_votes_),
    "min_max_mean" : sum(min_max_votes_) / len(min_max_votes_),
}
print("for popularity :")
print("- the min is :", round(pop["pop_min"], 2))
print("- the max is :", round(pop["pop_max"], 2))
print("- the mean is :", round(pop["pop_mean"], 2))
print("-"*30)
print("for vote count :")
print("- the min is :", round(pop["votes_min"], 2))
print("- the max is :", round(pop["votes_max"], 2))
print("- the mean is :", round(pop["votes_mean"], 2))
print("-"*30)
print("for vote average :")
print("- the min is :", round(pop["min_max_min"], 2))
print("- the max is :", round(pop["min_max_max"], 2))
print("- the mean is :", round(pop["min_max_mean"], 2))
print("-"*30)
```

what is the most popular movie for each year  between 2010 and 2023

solution print

image.png

```python
top_pop = {}
for x in data:
    date = x["release_date"][:4]
    if 2010 <= int(date) <= 2023:
        info = {
            "title": x["title"],
            "popularity": x["popularity"],
        }
        if date in top_pop:
            if info["popularity"] > top_pop[date]["popularity"]:
                top_pop[date] = info
        else:
            top_pop[date] = info

for k, v in sorted(top_pop.items(), key=lambda x: x[0], reverse=True):
    print(f"For {k} the most popular movies is {v}")
```
