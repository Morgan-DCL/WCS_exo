```python
import random

class DataAnalyst():

    def __init__(
        self,
        nom: str,
        prenom: str,
        age: int,
        sexe: str,
        prev_formation: str,
        motivation: int = 100,
        progress: int = 0,
    ):
        self.nom = nom
        self.prenom =prenom
        self.age = age
        self.sexe = sexe
        self.prev_formation = prev_formation
        self.motivation = motivation
        self.progress = progress

    def suivre_formation(self):
        """
        chaque fois qu'elle est utilisée, cette méthode diminue la motivation
        d'une valeur entière aléatoire (random) entre 5 et 25 et
        augmente la progression d'une valeur entière aléatoire entre 20 et 30
        """
        self.motivation -= random.randint(5, 25)
        self.progress += random.randint(20, 30)


    def bosser_plus(self):
        """
        chaque fois qu'elle est utilisée, cette méthode augmente la motivation et
        la progression d'une valeur entière aléatoire entre 10 et 30
        """
        self.motivation += random.randint(10, 30)
        self.progress += random.randint(10, 30)

    def echouer(self):
        """
        chaque fois qu'elle est utilisée baisse la motivation d'une valeur entre 20 et
        40 et augmente la progression d'une valeur entre 20 et 40
        """
        self.motivation -= random.randint(20, 40)
        self.progress += random.randint(20, 40)

    def reussir(self):
        """
        chaque fois qu'elle est utilisée, augmente la motivation d'une valeur entre 20 et
        40 et augmente la progression d'une valeur random entre 10 et 20
        """
        self.motivation += random.randint(20, 40)
        self.progress += random.randint(10, 20)

```

```python
def main(data_analyst):
    tour = 0
    funcs = [
        (data_analyst.suivre_formation, "Suivre Formation"),
        (data_analyst.bosser_plus, "Bosser Plus"),
        (data_analyst.echouer, "Echouer"),
        (data_analyst.reussir, "Reussir")
    ]

    while (
        data_analyst.motivation > 0 and
        data_analyst.progress < 100
        ):
        tour += 1
        func, func_name = random.choice(funcs)
        func()

        cmt = f"Tour {tour} \n"
        cmt += f"La méthode {func_name} a été utilisée.\n"
        cmt += f"Motivation vaut maintenant {data_analyst.motivation}\n"
        cmt += f"Progression vaut maitenant {data_analyst.progress}"
        print(cmt)
        print("-"*30)

        if data_analyst.motivation >= 100:
            return "BRAVO TU AS GAGNÉ !!!"
        elif data_analyst.progress >= 100:
            return "BRAVO TU AS APPRIS !!!"

```

```python
da = DataAnalyst(
    nom="Leitao",
    prenom="Morgan",
    age=36,
    sexe="Homme",
    prev_formation="Wild Code School"
)

cmt = main(da)
print(cmt)
```
